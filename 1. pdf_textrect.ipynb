```python
file_head = 'admin/city_council_minutes/Minutes_'
file_suffix = '.pdf'
target_head = 'min_'

d = dict()

for i in range(1,10):
    d[str(i)] = textract.process((file_head + str(0) + str(i) + file_suffix), method='pdfminer')

for i in range(10,46):
    d[str(i)] = textract.process((file_head + str(i) + file_suffix), method='pdfminer')
```

```python
def save_obj(obj, name ):
    with open('data/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('data/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
save_obj(d, "council_minutes_dict")
```
